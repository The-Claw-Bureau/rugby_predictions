```python
# Main function to run the entire scraping process
def scrape_competitions(competitions, years, driver):

    
    all_events_df = pd.DataFrame()
    
    # Loop through each competition URL
    for comp in competitions:
        
        print('')
        print('Scraping Competition: %s'%(comp))
        
        # Getting the competition name - Not sure why?
#         comp_name = extract_competition_name(comp)
        
        # Scrape the main competition page and subsequent pages (this should be the most recent year)
        all_events_df = scrape_competition_pages(all_events_df, driver, comp)
        print('len(all_events_df), comp, Main', len(all_events_df), comp)

        # Scrape past seasons
        for year in years:
            
            print('Trying previous  year:', year)
            all_events_df = scrape_past_seasons(all_events_df, driver, comp, year)
            print('len(all_events_df), comp, year', len(all_events_df), comp, year)
            
            
    return all_events_df


            
# Extract the competition name from the URL
def extract_competition_name(comp_link):
    comp_name = comp_link.replace('https://www.oddsportal.com/rugby-union/', '')
    return comp_name[:comp_name.find("/results")]


# Scrape the current competition's pages and gather event data
def scrape_competition_pages(all_events_df, driver, comp):
        
    
    # Navigate to the main competition page
    print('Trying: %s'%(comp))
    driver.get(comp)
    time.sleep(3)

    
    xpathselector = '//*[@id="app"]/div/div[1]/div/main/div[3]/div[4]'
    
    try:
        
        # Get the events from the first page
        all_events_df, new_events_df = get_events(all_events_df, driver, xpathselector, comp)
        
        print('Events Found: ', len(new_events_df), len(all_events_df))

        if len(new_events_df) == 0:
            return
        
        # Loop through the next 5 pages of the competition results
        for i in range(5):
            
            comp_link = comp + f'#/page/{i+2}/'
            print('Trying: %s'%(comp_link))
            driver.get(comp_link)
            driver.refresh()
            time.sleep(5)

            all_events_df, new_events_df = get_events(all_events_df, driver, xpathselector, comp)
            print('Events Found: ', len(new_events_df), len(all_events_df))
            if len(new_events_df) == 0:
                return
    
    except NoSuchElementException:
        pass

    finally:
        
        return all_events_df
    
    
    
# Get the events from the current page
def get_events(all_events_df, driver, xpathselector, comp):
    
    new_events_df = pd.DataFrame()
    
    try:
        
        events_table = driver.find_element(webdriver.common.by.By.XPATH, xpathselector)
        competition_name = ''.join(events_table.text.split('\n')[:5])
        events_list = events_table.find_elements(webdriver.common.by.By.CLASS_NAME, 'eventRow')

        # Retry to extract event IDs in case of an error
        event_ids = retry_extract_event_ids(events_list)
        event_ids = [x for x in event_ids if x != '']

        # Create a DataFrame for the current set of event IDs
        new_events_df = create_events_df(event_ids, competition_name, comp)
        print('Test - ' ,len(new_events_df), event_ids)
                
        if len(new_events_df) > 0:
            # Concatenate to the main DataFrame
            print(len(new_events_df), len(all_events_df))
            all_events_df = pd.concat([all_events_df, new_events_df], axis = 0, ignore_index=True)

#         driver.refresh()
    
    except NoSuchElementException:
        pass
    
    finally:
        
        return all_events_df, new_events_df

    
    
# Retry mechanism for extracting event IDs
def retry_extract_event_ids(events_list):
    while True:
        try:
            return [event.get_attribute(name='id') for event in events_list]
        except Exception:
            print("Retrying event extraction...")
            time.sleep(2)

            
# Create a DataFrame for the event information
def create_events_df(event_ids, competition_name, comp):
    
    events_df = pd.DataFrame(event_ids, columns=['external_event_id'])
    events_df['season_id'] = datetime.datetime.now().year
    events_df['competition_external_id'] = competition_name
    events_df['comp_link'] = comp

    # Remove rows where competition name contains 'Unfortunately'
    events_df = events_df[~events_df['competition_external_id'].str.contains('Unfortunately, ')]
    
    return events_df


# Scrape past seasons for the competition
def scrape_past_seasons(all_events_df, driver, comp, year):
    
    
    len_events = 0
    xpathselector = '//*[@id="app"]/div/div[1]/div/main/div[3]/div[4]'

    breakout_loop = False
    loop_num = 0
    
    while breakout_loop == False:
        
        loop_num += 1
        
        if loop_num >= 5:
            breakout_loop = True


        try:

            for loop in range(0,2):

                if loop == 0:
                    replacement_string = f'-{year}-{year+1}/results/'
                    comp_link = comp.replace('/results/', replacement_string)
                elif loop == 1:
                    replacement_string = f'-{year}/results/'
                    comp_link = comp.replace('/results/', replacement_string)


                print('Getting first page of previous season: ', comp_link)
                driver.get(comp_link)
                time.sleep(5)

                all_events_df, new_events_df = get_events(all_events_df, driver, xpathselector, comp)
                print('Events Found: ', len(new_events_df), len(all_events_df))

                # Break out of the season format loop
                if len(new_events_df) >  0:
                    breakout_loop = True
                    break

            # Break out of the retry page loop
            breakout_loop = True
            break


        except TimeoutException:

            continue

    
    if len(new_events_df) == 0:
        return all_events_df
    
    
    try:
        
        print('Looping through other results pages of this season')

        # Loop through pages for the past seasons
        for i in range(5):
            
            comp_link = comp.replace('/results/', replacement_string) + f'#/page/{i+2}/'
            print('Trying: ', comp_link)
            driver.get(comp_link)
            driver.refresh()
            time.sleep(4)
            all_events_df, new_events_df = get_events(all_events_df, driver, xpathselector, comp)
            print('Events Found: ', len(new_events_df), len(all_events_df))
            if len(new_events_df) == 0:
                return all_events_df

    except NoSuchElementException:
        return all_events_df
    
    
    finally:
        
        print('scrape_past_seasons - passing back all_events_df', len(all_events_df))
        return all_events_df
```

```python
# Function to get the longest valid event link from the list of links
def get_event_link(event_hrefs):
    """
    Extract the longest href link from the list of elements.
    Filters out None values.
    """
    all_links = [x.get_attribute('href') for x in event_hrefs if x.get_attribute('href') is not None]
    return max(all_links, key=len, default=None)  # Default to None if list is empty

# Function to generate a list of dates for the past 60 days and the next 30 days
def generate_date_list(num_days_prev, num_days_post):
    """
    Generates a list of dates, starting from 60 days before today to 30 days after today.
    """
    todays_date = datetime.date.today()
    date_list = [todays_date - datetime.timedelta(days=x) for x in range(num_days_prev)]
    date_list += [todays_date + datetime.timedelta(days=x) for x in range(num_days_post)]
    return date_list

# Function to navigate to a web link and handle timeouts
def navigate_to_webpage(driver, weblink, retries=3):
    """
    Attempts to navigate to a webpage. Retries multiple times in case of TimeoutException.
    """
    for _ in range(retries):
        try:
            driver.get(weblink)
            time.sleep(2)
            return True  # If navigation is successful
        except TimeoutException:
            time.sleep(1)
    return False  # If navigation fails after retries


    
# Function to process the event links and store them in the DataFrame
def process_event_links(event_links, all_events_df):
    """
    Processes a list of event links and appends relevant data to the global DataFrame.
    """
    
    # Parse the event links and construct a DataFrame
    events_df = pd.DataFrame([x.split('/') for x in event_links])
    
    # Ensure necessary columns exist and filter out invalid entries
    if len(events_df.columns) >= 7:
        
        events_df = events_df[(events_df[6].notna()) & (events_df[6] != '')]
        
        # Create required columns for the events DataFrame
        events_df['competition_external_id'] = events_df[[3,4,5]].apply(lambda x: '/'.join(x), axis=1)
        events_df['comp_link'] = events_df[range(6)].apply(lambda x: '/'.join(x), axis=1)
        events_df['external_event_id'] = events_df[6].apply(lambda x: x.split('-')[-1])
        
        # Keep only the necessary columns
        temp_df = events_df[['external_event_id', 'competition_external_id', 'comp_link']]
        
        # Append to the global DataFrame
        all_events_df = pd.concat([all_events_df, temp_df], sort=False, ignore_index=True)

        
        return all_events_df
    
    
# Main function to iterate through date list and scrape events
def scrape_recent_events(driver, num_days_prev, num_days_post):
    
    """
    Main function to iterate through recent event dates and scrape event data.
    """
    
    all_events_df = pd.DataFrame()
    
    date_list = generate_date_list(num_days_prev, num_days_post)  # Get the date list

    for sample_date in date_list:
        print(f"Processing date: {sample_date}")
        formatted_date = sample_date.strftime('%Y%m%d')
        
        # Construct the link for rugby events on the given date
        weblink = f'https://www.oddsportal.com/matches/rugby-union/{formatted_date}'
        
        # Navigate to the webpage
        if navigate_to_webpage(driver, weblink):
#             time.sleep(3)
            # If successful, scrape the event links
            all_events_df = scrape_event_links(all_events_df, driver, sample_date)
        else:
            print(f"Failed to load webpage for date: {formatted_date}")
            
            
    return all_events_df
```

```python
import pandas as pd
from selenium.common.exceptions import (
    ElementNotInteractableException, 
    NoSuchElementException, 
    StaleElementReferenceException, 
    TimeoutException,
    ElementClickInterceptedException
)
from IPython.display import clear_output


ha_lookup = {2: 'home', 3: 'draw', 4: 'away'}

# Function to retrieve team names, scores, and event details
def get_event_details(driver, comp_link, row_external_event_id):
    """
    Fetches event details like team names, scores, and event start time.
    """
    
    print('get_event_details')
    
    if comp_link[-1] == '/':
        event_link =  comp_link + row_external_event_id
    else:
        event_link =  comp_link + '/' + row_external_event_id

    for _ in range(3):  # Retry mechanism for loading the page
        try:
            print(event_link)
            driver.get(event_link)
            time.sleep(3)
            event_name_scores = driver.find_element(webdriver.common.by.By.XPATH, 
                '//*[@id="app"]/div[1]/div[1]/div/main/div[3]/div[2]/div[1]/div[1]')
            team_names = event_name_scores.text.split('\n')
            return team_names
        except NoSuchElementException:
            print('Trying to refresh')
            driver.refresh()
            time.sleep(2)
    
    return None

# Function to handle the extraction of team details
def extract_team_details(team_names):
    
    print('extract_team_details')
    print(team_names)
    """
    Extracts home and away team names and scores from the team names list.
    """
    try:
        return team_names[0], team_names[3], team_names[1], team_names[4]
    except IndexError:
        # Handle cases where data might be incomplete
        return team_names[0], team_names[2], '', ''

    
# Function to fetch event date and time
def get_event_datetime(driver):
    """
    Fetches the event start time.
    """
    
    print('get_event_datetime')
    
    try:
#         event_date_time = driver.find_element(webdriver.common.by.By.XPATH, 
#             '//*[@id="app"]/div[1]/div[1]/div/main/div[3]/div[2]/div[1]/div[2]/div[1]').text
        
        event_date_time = driver.find_element(webdriver.common.by.By.XPATH, '//*[@id="react-event-header"]/div/div/div[2]/div[1]/p[2]').text
        print(event_date_time)

#         event_date_time = event_date_time[event_date_time.find(',')+2:]
        event_date_time = event_date_time[:-1]
        # Clean up the date time string
#         event_date_time = event_date_time.replace('today,', '').replace('yesterday,', '').replace('tomorrow,', '')
        parsed_date = datetime.datetime.strptime(event_date_time, "%d %b %Y, %H:%M")
        formatted_date = parsed_date.strftime("%Y-%m-%d %H:%M:%S")

        return formatted_date

    except (NoSuchElementException, ValueError):
        return None

    
def get_event_datetime(driver):
    """
    Fetches the event start time and returns it in 'YYYY-MM-DD HH:MM:SS' format.
    Handles dates formatted as '19 Jan 2025,'.
    """
    print('get_event_datetime')
    
    try:
        event_date_time = driver.find_element(webdriver.common.by.By.XPATH, '//*[@id="react-event-header"]/div/div/div[2]/div[1]/p[2]').text
        print(event_date_time)
        
        # Remove trailing comma if present
        event_date_time = event_date_time.rstrip(',')
        
        # Attempt to parse different expected formats
        try:
            parsed_date = datetime.datetime.strptime(event_date_time, "%d %b %Y")
        except ValueError:
            try:
                parsed_date = datetime.datetime.strptime(event_date_time, "%d %b %Y, %H:%M")
            except ValueError:
                return None  # Return None if format is unexpected
        
        formatted_date = parsed_date.strftime("%Y-%m-%d %H:%M:%S")
        return formatted_date
    
    except NoSuchElementException:
        return None


# Function to fetch final and halftime scores
def get_final_and_halftime_scores(driver):
    """
    Fetches full-time and half-time scores.
    """
    
    print('get_final_and_halftime_scores')
    
    iteration_count = 0
    while iteration_count < 3:
        try:
            final_scores = driver.find_element(webdriver.common.by.By.XPATH, 
                '//*[@id="app"]/div[1]/div[1]/div/main/div[3]/div[2]/div[1]/div[2]/div[3]/div[2]').text.split('\n')
            return final_scores
        except Exception:
            print('Trying to get halftime results')
            driver.refresh()
            time.sleep(2)
            iteration_count += 1
    return None



def get_final_and_halftime_scores(driver):
    """
    Fetches full-time and half-time scores, and closes unexpected advertisement tabs.
    """
    
    print('get_final_and_halftime_scores')
    
    
    iteration_count = 0
    while iteration_count < 3:
        try:
            # Save the initial set of tabs
            original_tabs = driver.window_handles

            # Attempt to fetch the final scores
            final_scores = driver.find_element(webdriver.common.by.By.XPATH, 
                '//*[@id="app"]/div[1]/div[1]/div/main/div[3]/div[2]/div[1]/div[2]/div[3]/div[2]').text.split('\n')

            return final_scores

        except Exception:
            print('Trying to get halftime results')

            # Refresh the page to retry
            driver.refresh()
            time.sleep(2)

            # Check if a new tab opened
            current_tabs = driver.window_handles
            if len(current_tabs) > len(original_tabs):
                # Switch to the new tab
                new_tab = [tab for tab in current_tabs if tab not in original_tabs][0]
                driver.switch_to.window(new_tab)
                print('Closing unexpected tab...')
                driver.close()

                # Switch back to the original tab
                driver.switch_to.window(original_tabs[0])

            iteration_count += 1

    return None



# Function to process and store fixture data
def process_fixture_data(driver, comp_link, row_external_event_id, all_fixtre_data):
    """
    Processes fixture data like teams, scores, and event time and adds them to the global DataFrame.
    """
    
    print('Processing fixture data')
        
    team_names = get_event_details(driver, comp_link, row_external_event_id)
    if team_names:
        home_team_external_id, away_team_external_id, home_team_score, away_team_score = extract_team_details(team_names)
        event_start_time = get_event_datetime(driver)
        
        if pd.to_datetime(event_start_time) < datetime.datetime.utcnow():
            final_scores = get_final_and_halftime_scores(driver)
        else:
            final_scores = None
            
        if final_scores:
            home_team_score, away_team_score, home_halftime_score, away_halftime_score = extract_scores(final_scores)
        else:
            home_team_score, away_team_score, home_halftime_score, away_halftime_score = None, None, None, None

        current_comp_info = '/'.join(driver.current_url.split('/')[:-1])
        fixture_data = {
            'external_event_id': row_external_event_id,
            'comp_link': current_comp_info,
            'home_team_external_id': home_team_external_id,
            'home_team_score': home_team_score,
            'away_team_external_id': away_team_external_id,
            'away_team_score': away_team_score,
            'event_start_time': event_start_time,
            'home_halftime_score': home_halftime_score,
            'away_halftime_score': away_halftime_score,
        }
        
        # Append the fixture data to the global DataFrame
        all_fixtre_data = pd.concat([all_fixtre_data, pd.DataFrame([fixture_data])], sort=False, ignore_index=True)
    
    return all_fixtre_data

# Function to extract full-time and half-time scores
def extract_scores(final_scores):
    """
    Extracts full-time and half-time scores from the final scores list.
    """
    try:
        fulltime_scores = final_scores[1].strip()
        home_team_score = fulltime_scores.split(':')[0]
        away_team_score = fulltime_scores.split(':')[1]
    except IndexError:
        home_team_score = None
        away_team_score = None
    
    try:
        halftime_scores = final_scores[2].strip().replace('(', '').replace(')', '').split(',')[0]
        home_halftime_score = halftime_scores.split(':')[0]
        away_halftime_score = halftime_scores.split(':')[1]
    except IndexError:
        home_halftime_score = None
        away_halftime_score = None
    
    return home_team_score, away_team_score, home_halftime_score, away_halftime_score


# Function to retrieve odds data from the page
def get_odds_data(driver, row_external_event_id, fixture_date, odds_df):
    """
    Scrapes odds data for the given event.
    """
    
    print('get_odds_data')
        
    close_loop = 0
    for i in range(2, 100):
        if close_loop == 0:
            odds_provider_xpath = f'//*[@id="app"]/div[1]/div[1]/div/main/div[3]/div[2]/div[2]/div/div/div[1]/div[{i}]/div[1]'
            try:
                odds_provider_1X2_cell = driver.find_element(webdriver.common.by.By.XPATH, odds_provider_xpath)
                                
            except:
                break  # No more odds providers available

            odds_provider = odds_provider_1X2_cell.text.split('\n')[0]
#             print('odds_provider', odds_provider)
            if odds_provider in ['Average', 'Highest', 'My coupon']:
                print('closing loop')
                close_loop = 1
            else:
                odds_df = scrape_odds_for_provider(driver, row_external_event_id, i, odds_provider, odds_df, fixture_date)
                

    # Now get handicap odds
    handicap_dicts = get_handicap_odds('Asian Handicap')
    if len(handicap_dicts) > 0:
        handicap_df = pd.DataFrame(handicap_dicts)
        handicap_df['external_event_id'] = row_external_event_id

    else:
        handicap_df = None

        
    # Now get over under odds
    over_under_dicts = get_handicap_odds('Over/Under')
    if len(over_under_dicts) > 0:
        over_under_df = pd.DataFrame(over_under_dicts)
        over_under_df['external_event_id'] = row_external_event_id

    else:
        over_under_df = None

    
    return odds_df, handicap_df, over_under_df



# Function to add the correct year
def add_year(row):
    # Extract day and month
    day_month = row['time']
    fixture_date = row['fixture_date']
    
    # Parse 'time' as a date without a year
    #             'time': [pd.to_datetime(time, format='%d %b, %H:%M') if time else None for time in first_row_text],
    
    if day_month is not None:
        
        if day_month[:6] == '29 Feb':
            
            target_date = pd.to_datetime(str(fixture_date.year) +'/02/29 ' + day_month[-5:])
            return target_date

        else:

            target_date = datetime.datetime.strptime(day_month, '%d %b, %H:%M').replace(year=fixture_date.year)

            # Cycle back year by year until the date is before the fixture date
            while target_date > fixture_date:
                target_date = target_date.replace(year=target_date.year - 1)

            # Return the corrected date as 'YYYY-MM-DD'
            return target_date

    else:
        
        return None


# Function to scrape odds for a specific provider
def scrape_odds_for_provider(driver, row_external_event_id, i, odds_provider, odds_df, fixture_date):
    """
    Scrapes odds data for a specific provider on the event page.
    
    """
#     j is looping home draw away
#     i is looping down odds providers
    
    for j in range(2, 5):
        odds_xpath = f'//*[@id="app"]/div[1]/div[1]/div/main/div[3]/div[2]/div[2]/div/div/div[1]/div[{i}]/div[{j}]/div'
        try:
            odds_table1X2_cell = driver.find_element(webdriver.common.by.By.XPATH, odds_xpath)
            
            if pd.to_datetime(fixture_date) > datetime.datetime.utcnow():
                # Create an ActionChains object
                actions = ActionChains(driver)
                # Hover over the element
                actions.move_to_element(odds_table1X2_cell).perform()
            else:
                odds_table1X2_cell.click()
                time.sleep(0.25)
            
            # Retrieve odds values and times
            temp_odds_df = get_odds_values(driver, i, j)
            
            if len(temp_odds_df) > 0:
                temp_odds_df = temp_odds_df[ pd.notna(temp_odds_df['odds'])]
            
            if len(temp_odds_df) > 0:
#                 temp_odds_df = pd.DataFrame(list(zip(odds_times, odds_values)), columns=['time', 'odds'])
                temp_odds_df['external_event_id'] = row_external_event_id
                temp_odds_df['home_away'] = ha_lookup[j]
                temp_odds_df['odds_provider'] = odds_provider
                temp_odds_df['fixture_date'] = pd.to_datetime(fixture_date, format='%Y-%m-%d')
                temp_odds_df['corrected_time'] = temp_odds_df.apply(add_year, axis=1)

                odds_df = pd.concat([odds_df, temp_odds_df], ignore_index=True)
                
#                 driver.refresh()
#                 time.sleep(1)
        except (ElementNotInteractableException, NoSuchElementException, ElementClickInterceptedException):
            pass
    
    return odds_df


# Function to get odds values and times
def get_odds_values(driver, i, j):
    
    """
    Fetches the odds values and times from the odds table.
    
    """

    # Times
    first_row_xpath = f'//*[@id="app"]/div[1]/div[1]/div/main/div[3]/div[2]/div[2]/div/div/div[1]/div[{i}]/div[{j}]/div/div[1]/div/div[1]/div/div[1]'
    # Odds
    second_row_xpath = f'//*[@id="app"]/div[1]/div[1]/div/main/div[3]/div[2]/div[2]/div/div/div[1]/div[{i}]/div[{j}]/div/div[1]/div/div[1]/div/div[2]' 
    
    
    first_row_text = driver.find_element(webdriver.common.by.By.XPATH, first_row_xpath).text.split('\n')
    second_row_text = driver.find_element(webdriver.common.by.By.XPATH, second_row_xpath).text.split('\n')

#     first_row_text = ''
#     second_row_text = ''
    
#     first_row_elements = driver.find_elements(webdriver.common.by.By.XPATH, first_row_xpath)
#     if first_row_elements:
#         # Element exists, access it with elements[0]
#         first_row_text = first_row_elements[0].text.split('\n')
#         print(first_row_text)
        
#     second_row_elements = driver.find_elements(webdriver.common.by.By.XPATH, second_row_xpath)
#     if second_row_elements:
#         # Element exists, access it with elements[0]
#         second_row_text = second_row_elements[0].text.split('\n')
#         print(second_row_text)

    
    if (len(first_row_text) > 0) & (len(second_row_text) > 0):
        
        # Create DataFrame with two columns: 'time' and 'odds'
        temp_odds_df = pd.DataFrame({
#             'time': [pd.to_datetime(time, format='%d %b, %H:%M') if time else None for time in first_row_text],
            'time': [time if time else None for time in first_row_text],
            'odds': pd.to_numeric(second_row_text, errors='coerce')
        })
        
        
        return temp_odds_df

    else:
        
        return None
    

# Main function to scrape all data for each event
def scrape_event_data(driver, all_events_df, get_odds_data_bool):
    """
    Iterates over all events and scrapes data like fixture details and odds.
    """
    
    print('scrape_event_data')
    
    all_handicap_df = pd.DataFrame()
    all_over_under_df = pd.DataFrame()
    odds_df = pd.DataFrame()
    all_fixtre_data = pd.DataFrame()
    

    loop_num = 0
    
    for row in all_events_df.index:
        
        loop_num += 1
        
        print(loop_num, len(all_events_df))
        
        row_external_event_id = all_events_df.loc[row, 'external_event_id']
        comp_link = all_events_df.loc[row, 'comp_link']

#         row_external_event_id = row[0]
#         comp_link = row[3].replace('results/','')
        all_fixtre_data = process_fixture_data(driver, comp_link, row_external_event_id, all_fixtre_data)
        
        fixture_date = all_fixtre_data[ all_fixtre_data['external_event_id'] == row_external_event_id]
        if len(fixture_date) > 0:
            fixture_date = fixture_date.iloc[0]['event_start_time']

            if get_odds_data_bool:

                odds_df, handicap_df, over_under_df = get_odds_data(driver, row_external_event_id, fixture_date, odds_df)
                

#                 if temp_odds_df is not None:
                print('Num odds retrieved: %s'%(len(odds_df)))
#                     odds_df = pd.concat([odds_df, temp_odds_df], ignore_index = True)
#                     odds_df.to_csv('odds_df.csv', index = False)
                
                if handicap_df is not None:
                    print('Num handicap_df retrieved: %s'%(len(handicap_df)))
                    handicap_df['time'] = fixture_date
                    all_handicap_df = pd.concat([all_handicap_df, handicap_df], ignore_index = True)
                    all_handicap_df.to_csv('all_handicap_df.csv', index = False)

                if over_under_df is not None:
                    print('Num over_under_df retrieved: %s'%(len(over_under_df)))
                    over_under_df['time'] = fixture_date
                    all_over_under_df = pd.concat([all_over_under_df, over_under_df], ignore_index = True)
                    all_over_under_df.to_csv('all_over_under_df.csv', index = False)

        
        # Clear output for better readability in notebooks
#         clear_output()
        print(f"Processed event: {row_external_event_id}")
    
    
    if len(odds_df) > 0:
        odds_df['time'] = odds_df['corrected_time']
    
    
    
    return all_fixtre_data, odds_df, all_handicap_df, all_over_under_df
```
